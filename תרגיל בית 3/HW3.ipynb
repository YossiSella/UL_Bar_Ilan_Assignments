```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Step 1: Generate synthetic dataset with 3 clusters
np.random.seed(42)
X, y_true = make_blobs(n_samples=300, centers=[(2, 2), (8, 8), (4, 0)], cluster_std=1.2)

# Step 2: Run K-means with random initialization (default)
kmeans_random = KMeans(n_clusters=3, init='random', random_state=200, n_init=1)
kmeans_random.fit(X)
labels_random = kmeans_random.labels_

# Step 3: Run K-means with custom initialization (manually choosing bad centroids)
init_centroids = np.array([[0, 0], [10, 10], [10, 0]])  # Poorly chosen centroids
kmeans_custom = KMeans(n_clusters=3, init=init_centroids, n_init=1, random_state=200)
kmeans_custom.fit(X)
labels_custom = kmeans_custom.labels_

# Step 4: Plot ground truth and K-means results
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Ground Truth Labels
ax[0].scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', s=20)
ax[0].set_title("Ground Truth Clusters")

# K-means with Random Initialization
ax[1].scatter(X[:, 0], X[:, 1], c=labels_random, cmap='viridis', s=20)
ax[1].scatter(kmeans_random.cluster_centers_[:, 0], kmeans_random.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
ax[1].set_title("K-means with Poor Initialization")
ax[1].legend()

# K-means with specific Initialization
ax[2].scatter(X[:, 0], X[:, 1], c=labels_custom, cmap='viridis', s=20)
ax[2].scatter(kmeans_custom.cluster_centers_[:, 0], kmeans_custom.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
ax[2].set_title("K-means with Good Initialization")
ax[2].legend()

plt.show()
```
